성별 데이터 저장하기 (Type 1)

```python
import csv

#파일 읽어오기
f = open('/content/201902_201902_연령별인구현황_월간_gender.csv')
data = csv.reader(f)

# 남성, 여성 데이터 감을 list 변수 선언 
m_list = []
f_list = []

# 조회할 동네명 입력 받기
address = input('검색할 동을 입력하세요 : ')

#남성 여성 인구수 데이터 읽어오기
for row in data: 
  if address in row[0]:
    for i in range(0,101):
      m_list.append(int(row[i + 3].replace(',',''))) #남성 데이터 3 ~ 103 index (0 ~ 100)
      f_list.append(int(row[-(i + 1)].replace(',',''))) # - 1부터 디텍션 해라. 여성 인덱스 -1 ~ -101 index (100 ~ 0)

print(f_list)

#거꾸로 수집된 여성의 연령 데이터를 뒤집기
f_list.reverse()

print(f_list)
print(m_list)
print(len(f_list))
print(len(m_list))
```

성별 데이터 저장하기 (Type 2)

```python
import csv

#파일 읽어오기
f = open('/content/201902_201902_연령별인구현황_월간_gender.csv')
data = csv.reader(f)

# 남성, 여성 데이터 감을 list 변수 선언 
m_list = []
f_list = []

# 조회할 동네명 입력 받기
address = input('검색할 동을 입력하세요 : ')

#남성 여성 인구수 데이터 읽어오기
for row in data: 
  if address in row[0]:
    
    #남성 데이터 수집
    for i in row[3:104]:
      m_list.append(int(i.replace(',','')))

    #여성 데이터 수집    
    for i in row[106:]:
      f_list.append(int(i.replace(',','')))

print(f_list)
print(m_list)
print(len(f_list))
print(len(m_list))
```

항이리 그래프를 위해 남성 데이터(연령수) 음수로 바꾸기

```python
import csv
import matplotlib.pyplot as plt

#파일 읽어오기
f = open('/content/201902_201902_연령별인구현황_월간_gender.csv')
data = csv.reader(f)

# 남성, 여성 데이터 감을 list 변수 선언 
m_list = []
f_list = []

# 조회할 동네명 입력 받기
address = input('검색할 동을 입력하세요 : ')

#남성 여성 인구수 데이터 읽어오기
for row in data: 
  if address in row[0]:
    
    #남성 데이터 수집
    for i in row[3:104]:
      m_list.append(-int(i.replace(',','')))

    #여성 데이터 수집    
    for i in row[106:]:
      f_list.append(int(i.replace(',','')))

#시각화 

#항아리 차트에서는 모양이 중요함

plt.figure(dpi = 200)
plt.title(address + '지역의 남녀 성별 인구 분포')
plt.barh(range(101), m_list, label = '남성')
plt.barh(range(101), f_list, label = '여성')
plt.gca().axes.xaxis.set_visible(False) #x축 지우기
plt.legend()
plt.show()
```

우리 동네 남녀 성별 비율 확인하기 -> 파이차트

```python
import csv
import matplotlib.pyplot as plt

#파일 읽어오기
f = open('/content/201902_201902_연령별인구현황_월간_gender.csv')
data = csv.reader(f)

# 남성, 여성 데이터 감을 list 변수 선언 
m_list = []
f_list = []

#남성, 여성 전체 데이터를 담을 list 변수 선언
total_list = []

# 조회할 동네명 입력 받기
address = input('검색할 동을 입력하세요 : ')

#남성 여성 인구수 데이터 읽어오기
for row in data: 
  if address in row[0]:

    #남성의 총인구수
    total_list.append(int(row[1].replace(',','')))
    
    #여성의 총인구
    total_list.append(int(row[104].replace(',','')))

     #남성 데이터 수집
    for i in row[3:104]:
      m_list.append(-int(i.replace(',','')))

    #여성 데이터 수집    
    for i in row[106:]:
      f_list.append(int(i.replace(',','')))

    break

#파이 차트 시각화
label = ['남','여']
plt.title(address + '지역의 남녀 성별 인구 비율')
plt.pie(total_list, labels = label, autopct = '%.2f%%', startangle= 90, explode = (0,0.1))
plt.legend()
plt.show()

#항아리 모양
plt.figure(dpi = 200)
plt.title(address + '지역의 남녀 성별 인구 분포')
plt.barh(range(101), m_list, label = '남성')
plt.barh(range(101), f_list, label = '여성')
plt.gca().axes.xaxis.set_visible(False) #x축 지우기
plt.legend()
plt.show()
```

파이차트 다양하게 그려보기

```python
size = [1244,2312,1031,1233]
label = ['A형', 'B형', 'AB형', 'O형']

plt.axis('equal') #없으면 원에 붙어 있음
# 특정 레이블을 표출해야 하는 경우
plt.pie(size, labels = label, autopct = '%.2f%%', explode = (0,0,0.1,0), shadow = True)  #startangle =  을 쓰지 않으면 비율이 큰 것 부터 오른쪽으로 표시
plt.legend()
plt.show()
```

bar 차트를 활용한 남녀 성별 인구 분포 

```python
import csv
import matplotlib.pyplot as plt

#파일 읽어오기
f = open('/content/201902_201902_연령별인구현황_월간_gender.csv')
data = csv.reader(f)

# 남성 연령 데이터에서 여성 연령 데이터를 뺀 값을 담을 list변수 (남성 인구수 - 여성 인구수)
result = []

#사용자 입력
address = input('검색할 동을 입력하세요 : ')

# 인구수 데이터 읽어오기
for row in data:
  if address in row[0]:
    for i in range(0,101):
      m_cnt = int(row[i + 3].replace(',', '')) #0세 ~ 100세 남성 인구수 (index 3부터 시작)
      f_cnt = int(row[i + 106].replace(',', '')) # #0세 ~ 100세 여성 인구수 (index 106부터 시작)
      result.append(m_cnt - f_cnt)
    break

#시각화
plt.figure(dpi = 100)
plt.title(address + ' 지역의 남녀 인구수 분포')
plt.bar(range(101), result)
plt.show()
```
