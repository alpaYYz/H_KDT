# 교통사고 데이터 전처리

# 지도 활용 하여 시각화

```python
import folium

map = folium.Map(location = [35.1595454,126.8526012]) # [경도, 위도]
map
```

```python
gj_df.head()
```

```python
gj_df = gj_df.astype({'사망자수':'float'})
gj_df = gj_df.astype({'부상자수':'float'})

gj_df.info()
```

```python
map = folium.Map(location = [35.1595454,126.8526012]) # 위도, 경도

# for문을 사용하여 gj_df 데이터가 있을 때까지 반복적으로 CircleMaker 지도에 add 작업 -> 74번 

for n in gj_df.index:
  cnt = gj_df['사망자수'][n] + gj_df['부상자수'][n]

  #위도 경도 추출
  lat = gj_df['위도'][n]#위도
  lng = gj_df['경도'][n]#경도

  folium.CircleMarker([lat, lng], radius = cnt * 10, popup = gj_df['사고유형'][n],
                      color = '#3186cc', fill_color = '#3186cc').add_to(map)

map
```

```python
map.save('2021_광주교통사고현황.html')
```
