```python
import numpy as np
import pandas as pd
```

```python
# 출력을 소수점 이하 3자리로 제한
%precision 3

# DataFrame 의 출력을 소수점 이하 3자리로 제한
#pd.set_option("display.precision", 3)
```

```python
df = pd.read_csv('/content/ch2_scores_em.csv')
df.head()
```

```python
scores = np.array(df['english'])[:10]
scores 
```

```python
scores_df = pd.DataFrame({'score' : scores},
                         index = pd.Index(['A', 'B', 'C', 'D', 'E', 
                                           'F', 'G', 'H', 'I', 'J'],
                                          name = 'student'))
scores_df
```

```python
sum(scores) / len(scores)
```

```python
np.mean(scores) # 평균
```

```python
scores_df.mean()
```

```python
sorted_scores = np.sort(scores)
sorted_scores
```

```python
n = len(sorted_scores)
if n % 2 == 0:
  m0 = sorted_scores[n//2 - 1]
  m1 = sorted_scores[n//2]
  median = (m0 + m1) / 2
else:
  median = sorted_scores[(n+1)//2 - 1]

median
```

```python
np.median(scores) # 중앙값
scores
```

```python
mean = np.mean(scores)
deviation = scores - mean
```

```python
deviation # 편차
```

```python
another_scores = [50, 60, 58, 54, 51, 56, 57, 53, 52, 59]
another_mean = np.mean(another_scores)
another_deviation = another_scores - another_mean
another_deviation
```

```python
np.mean(deviation) # 편차의 평균
```

```python
np.mean(another_deviation) 
```

```python
summary_df = scores_df.copy()
summary_df['deviation'] = deviation
summary_df
```

```python
summary_df.mean()
```

```python
np.var(scores) == np.mean(deviation ** 2) # 분산
```

```python
summary_df['square of deviation'] = np.square(deviation)
summary_df
```

```python
summary_df.mean()
```
