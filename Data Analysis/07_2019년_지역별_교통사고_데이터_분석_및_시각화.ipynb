```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib as plt
import koreanize_matplotlib
```

데이터 준비하기

```python
# https://www.data.go.kr/data/15070340/fileData.do -> 교통사고 정보

df = pd.read_csv('/content/도로교통공단_교통사고 정보_2019.csv', encoding = 'euc-kr')
df.head()
```

```python
df.info()
```

사망교통사고 데이터 전처리

```python
# 사망사고가 가장많이 발생하는 시간은?
# 발생년월일시 -> 뒤에 두자리  (시간 = 00 ~ 23)
# 2019010100 -> 00 분리 -> 분리 시킬 수 있는 keyword가 없음 -> 문자열 슬라이딩

# 1) 발생년월일시 연속적인 숫자(int) -> 문자열 슬라이딩 할 수 있는 string 형변환

df = df.astype({'발생년월일시' : 'string'})
df.head()
```

```python
# 2) 시간으로 분리한 데이터를 발생시간 컬럼으로 담기
df['발생시간'] = df['발생년월일시'].str[8:]
df.head()
```

```python
df.info()
```

```python
# 3) 발생시간 컬럼을 숫자 데이터타입으로 변경
df = df.astype({'발생시간' : 'int'})
df.info()
```

```python
df['발생년월일시'].str[:8]
```

```python
# 4) 발생년월일시(string) -> 날짜 타입으로 변경
# ex) 2019010100 -> 20190101(년월일) 슬라이딩 -> YYYY-MM-DD (날짜 형태로 변경)
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8], format = '%Y-%m-%d', errors = 'raise') # y는 대문자로 해야함.
df.info()
```

지도를 활용하여 교통사고 현황 분석

```python
import folium

map = folium.Map(location = [36.321655, 127.378953]) # [경도, 위도]
map
```

```python
# 사망사고 위치에 CircleMarker 표출
# CircleMarker 사이즈 = 사망자수 + 부상자수(부상자수 = 중상자수 + 경상자수 + 부상신고자수)
# 클릭하면 -> popup 사고유형 나타내도록 구현

# CircleMarker 사이즈(radius)는 실수형 데이터 타입만 허용
# 사망자수, 부상자수 컬럼을 -> 실수형 데이터 타입으로 변경
dj_df = dj_df.astype({'사망자수':'float'})
dj_df = dj_df.astype({'부상자수':'float'})

dj_df.info()
```

```python
dj_df.head()
```

사망자수 + 부상자수에 따른 CircleMaker 지도에 표출하기

```python
map = folium.Map(location = [36.321655, 127.378953]) # 위도, 경도

# for문을 사용하여 dj_df 데이터가 있을 때까지 반복적으로 CircleMaker 지도에 add 작업 -> 74번 

for n in dj_df.index:
  cnt = dj_df['사망자수'][n] + dj_df['부상자수'][n]

  #위도 경도 추출
  lat = dj_df['위도'][n]#위도
  lng = dj_df['경도'][n]#경도

  folium.CircleMarker([lat, lng], radius = cnt * 10, popup = dj_df['사고유형'][n],
                      color = '#3186cc', fill_color = '#3186cc').add_to(map)

map
```

```python
map.save('2019_대전교통사고현황.html')
```
