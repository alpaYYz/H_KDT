우리동네 연령별 인구수 result 리스트에 저장 -> 시각화를 위해

```python
import csv
# 1) 파일 읽기

f = open('/content/201902_201902_연령별인구현황_월간 (2).csv')
data = csv.reader(f)

# 2) 시각를 위한 리스트 변수 선언
result = []

# 3) 데이터 한줄 씩 읽기
for row in data:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if '관저1동' in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result.append(int(i)) #문자열을 정수 형변환 ->숫자형을 변환함

# 4) 데이터 확인 
print(result)
print(len(result))
```

사용자 입력(시, 동)을 입력을 통한 인구 구조 시각화 하기

```python
import csv
import matplotlib.pyplot as plt
# 1) 파일 읽기

f = open('/content/201902_201902_연령별인구현황_월간 (2).csv')
data = csv.reader(f)

# 2) 시각를 위한 리스트 변수 선언
result = []

# 3) 사용자 입력 받기
si = input('시(광역시) 입력하세요 : ')
dong = input('동네명(동)을 입력하세요 :')

# 4) 데이터 한줄 씩 읽기
for row in data:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if si in row[0] and dong in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함


# 5) 시각화 하기

plt.figure(dpi = 100)
plt.style.use('ggplot') #격자 무늬 스타일 지정
plt.plot(result)
plt.title(si + ' ' + dong + "의 인구 구조 그래프")
plt.show()
```

두개의 동네를 입력받아 인구구조 비교하기

```python
import csv
import matplotlib.pyplot as plt
# 1) 파일 읽기

f = open('/content/201902_201902_연령별인구현황_월간 (2).csv')
data = csv.reader(f)

# 2) 시각를 위한 리스트 변수 선언
result1 = []
result2 = []

# 3) 사용자 입력 받기
ad1 = input('우리 동네(동) 입력하세요 : ')
ad2 = input('비교할 동네(동) 입력하세요 :')

# 4) 데이터 한줄 씩 읽기
for row in data:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad1 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result1.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함

  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad2 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result2.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함


# 5) 시각화 하기

plt.figure(dpi = 100)
plt.style.use('ggplot') #격자 무늬 스타일 지정
plt.title(ad1 + ' ' + 'vs' + ' ' + ad2 + "의 인구 구조 비교 그래프")
plt.plot(result1, 'red', label = ad1)
plt.plot(result2, 'blue', label = ad2)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
# 1) 파일 읽기
# 안되면 encoding='utf-8-sig'
f1 = open('/content/201902_201902_연령별인구현황_월간 (2).csv')
f2 = open('/content/202002_202002_연령별인구현황_월간 (2).csv')
f3 = open('/content/202102_202102_연령별인구현황_월간 (1).csv')
f4 = open('/content/202202_202202_연령별인구현황_월간 (1).csv')

data1 = csv.reader(f1)
data2 = csv.reader(f2)
data3 = csv.reader(f3)
data4 = csv.reader(f4)

# 2) 시각를 위한 리스트 변수 선언
result1 = []
result2 = []
result3 = []
result4 = []


# 3) 사용자 입력 받기
ad1 = input('시를 입력하세요 : ')
ad2 = input('동을 입력하세요 : ')

# 4) 데이터 한줄 씩 읽기
for row in data1:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad1 in row[0] and ad2 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result1.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함

for row in data2:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad1 in row[0] and ad2 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result2.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함

for row in data3:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad1 in row[0] and ad2 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result3.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함

for row in data4:
  #신도림 데이터만 추출 -> 행정구역은 row[0]에 들어 있음
  if ad1 in row[0] and ad2 in row[0]:
    #라스트 변수에 담기
    for i in row[3:]:
      result4.append(int(i.replace(',',''))) #문자열을 정수 형변환 ->숫자형을 변환함


# 5) 시각화 하기

plt.figure(dpi = 100)
plt.style.use('ggplot') #격자 무늬 스타일 지정
plt.title(ad1 + ' ' + ad2 + '의 ' + "의 4년간 인구 구조 비교 그래프")
plt.plot(result1, 'red', label = '2019년' + ad2)
plt.plot(result2, 'blue', label = '2020년' + ad2)
plt.plot(result3, 'green', label = '2021년' + ad2)
plt.plot(result4, 'yellow', label = '2022년' + ad2)
plt.legend()
plt.show()

#결과 저장하기.
#plt.savefig(ad1 + ' ' + ad2 + '의' + "의 2019 ~ 2021(4년간) 인구 구조 비교 그래프.png", dpi = 200, format = 'png')
```
