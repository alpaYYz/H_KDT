```python
# 학교 합치기 + 취업율 평균
college_df_employee = df.groupby(['학교명'],as_index=False).agg(취업률 = ('취업률(%)\n[B/{A-(C+D+E+F+G)x100','mean')).sort_values('취업률',ascending=False)
college_df_employee
```
