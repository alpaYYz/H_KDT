```python
import pandas as pd

df = pd.read_csv('/content/201902_201902_연령별인구현황_월간.csv')
df.head(10)
```

```python
df.info()
```

데이터 전처리(정제)

```python
# 1) 행정구역의 컬럼의 행정 코드를 삭제하고 주소 데이터만 사용
# ex) 서울특별시 종로구 가회동(1111060000) -> 서울특별시 종로구 가회동

df['행정구역'] = df['행정구역'].str.split('(').str.get(0)
df.head()
```

```python
# 2) 행정구역을 제외한 컬럼명 변경
# ex) 2019년02월_계_총인구수 -> 총인구수, 2019년02월_계_0세 -> 0세
colums_list = df.columns # 기존 컬럼 이름이 담긴 list

final_col = {} # {key : value} -> {기존컬럼명 : 바꿀 컬럼명} -> {2019년02월_계_0세 : 0세} 딕셔너리 

for col in colums_list:
  final_col[col] = col.split('_')[-1]

# 컬럼명 변경
df.rename(columns = final_col, inplace = True)
df.head()
```

```python
# 3) 데이터 콤마(,) 제거
# ex) 9,770,638	 -> 9770638 -> int형 변환 (5번에서 처리 예정)	

#df.replace(',','', inplace = True)

# 특수기호를 모두 제거하는 정규식을 replace -> 모든 특수 기호를 처리하는 방법 -> 정규식 (regular expression)
df.replace('[^\w]','', regex =True, inplace = True) #특수기호 전부 지워주는 정규식
```

```python
# 4) 인덱스 변경 0,1,2,3 -> 행정구역명을 인덱스로 변경  
# ex) 0 -> 서울특별시
df.set_index(keys = ['행정구역'], inplace = True)
```

```python
# 5) 형변환 
df = df.astype('int')
df.info()
```

```python
df.head()
```

```python
# 6) 전체 데이터를 총 인구수로 나눠서 비율로 계산
df = df.div(df['총인구수'], axis = 0)
df.head()
```

```python
# 7) 필요없는 컬럼 삭제 -> 총인구수, 연령구간인구수 컬럼 삭제

del df['총인구수'], df['연령구간인구수']
```

```python
df.head()
```

```python
# 8) 전처리(정제)한 파일 저장
df.to_csv('201902_인구구조_전처리데이터.csv', encoding = 'utf-8-sig')
```

궁금한 지역의 이름을 받고 해당하는 인구구조 저장하기

```python
address = input('인구구조가 알고 싶은지역의 주소를 입력하시오 : ')

cons = df.index.str.contains(address) #해당 지역의 인덱스찾기 -> contains() in 연산자와 동일
df2 = df[cons]
df2.head()
```

```python
# 사용자가 선택한 지역 인구구조 시각화
import matplotlib.pyplot as plt

plt.style.use('ggplot')
df2.T.plot()
plt.show
```

사용자가 선택한 인구 구좌와 비슷한 동네 찾기 TOP 5

```python
import numpy as np

# 전체데이터 -> 잠실2동의 연령별 인구 비율 빼기 -> 제곱 시키기(모두 양수로 변환) -> 총 합계 (sum()) -> top5 (자신 제외)

top5 = np.power(df.sub(df2.iloc[0], axis = 1),2).sum(axis = 1).sort_values().index[1:6]
top5
```

```python
# 모든 지역 - 인구구조 빼기 (각 컬럼별로 빼기)
df.sub(df2.iloc[0], axis = 1)
```

```python
# 제곱 양수 만들기
np.power(df.sub(df2.iloc[0], axis = 1),2)
```

```python
# 각 지역별 인구 비율의 총 합계
np.power(df.sub(df2.iloc[0], axis = 1),2).sum(axis = 1)
```

```python
# 오름차순으로 줄 세우기
np.power(df.sub(df2.iloc[0], axis = 1),2).sum(axis = 1).sort_values()
```

```python
#비교 지역을 제외한 top5 뽑기
np.power(df.sub(df2.iloc[0], axis = 1),2).sum(axis = 1).sort_values().index[1:6]
```

사용자가 선택한 인구구조와 비슷한 5곳 인구구조 시각화

```python

plt.style.use('ggplot')
ax = df.loc[np.power(df.sub(df2.iloc[0], axis = 1), 2).sum(axis = 1).sort_values().index[1:6]].T.plot()
plt.title(address + ' 와 비슷한 인구구조 TOP 5')
#plt.show()

# 5) 우리동네와 비슷한 인구구조 그래프 저장
ax.figure.savefig(address + ' 와 비슷한 인구구조 TOP5.pdf')
```
