```python
import pandas as pd
import re

import konlpy

import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
muh = open('/content/노무현대통령_연설문.txt').read()
muh 
```

```python
muh = re.sub('[^가-힣]', ' ', muh)
muh
```

```python
hannanum = konlpy.tag.Hannanum()
h_muh_nn = hannanum.nouns(muh)
h_muh_nn
```

```python
# 확인
komoran = konlpy.tag.Komoran()
k_muh_nn = komoran.nouns(muh)
k_muh_nn
```

```python
# 데이터프레임 변환
k_muword_df = pd.DataFrame({'word' : k_muh_nn})
k_muword_df
```

```python
# 빈도표 만들기
k_muword_df['count'] = k_muword_df['word'].str.len()
k_muword_df
```

```python
# 중복 단어 삭제
k_muword_df = k_muword_df.query('count >= 2') # 두글자 이상만 출력
k_muword_df
```

```python
k_muword_df = k_muword_df.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
k_muword_df
```

```python
k_mutop20 = k_muword_df.head(20)
```

```python
fig, axes = plt.subplots(figsize = (15,5)) # (컬럼의 갯수, 로우의 갯수)
plt.suptitle('노무현 대통령 취임식 단어 분석')

sns.barplot(data = k_mutop20, x = 'n', y = 'word')
axes.set_title('Komoran')

plt.show()
```

```python
font_path = '/content/BMDOHYEON_ttf.ttf'
```

```python
k_dic_word = k_muword_df.set_index('word').to_dict()['n'] # 데이터 프레임을 딕셔너리로 만들었음
k_dic_word
```

```python
from wordcloud import WordCloud

# 워드 클라우드 만들기 
wc = WordCloud(width = 1000, height = 700, font_path = font_path)
img_wordcloud = wc.generate_from_frequencies(k_dic_word)

plt.subplots(figsize = (25,15))
plt.axis('off')
plt.imshow(img_wordcloud, interpolation = 'bilinear')
plt.show()
```
