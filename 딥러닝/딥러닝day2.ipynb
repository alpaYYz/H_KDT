```python
z = torch.tensor([[1,2,3],
                  [4,5,6]])
z.shape
```

```python
torch.sum(z, dim = 0) # dim = 0 은 열 방향 세로 방향을 다 더한 후 1차원 배열을 만든다 1은 행 가로 방형으로 다 더한 후 1차원 배열로 나옴
```

```python
torch.max(torch.tensor([1,2,3]), dim = 0) # 여기서의 dim은 벨류를 주고 거기의 인덱스를 같이 준다.
```

```python
x = torch.tensor([1,2,3])
value, index = torch.max(x, dim = 0)
print(value) # 최대값은 3
print(index) # 최대값의 인덱스의 위치는 2
```

```python
torch.argmax(x, dim = 0) # argmax -> 최대값의 인덱스를 출력함
```
