# 2. Multiple Linear Regression

원 - 핫 인코딩

```python
import pandas as pd
dataset = pd.read_csv('/content/MultipleLinearRegressionData.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(drop = 'first'), [2])], remainder = 'passthrough') # [2]는 두번쨰 인덱스에서 함, 함수 
X = ct.fit_transform(X)
X

# 1.0 : home
# 0.1 : library 
# 0.0 : cafe
```

### 데이터 세트 분리

```python
, train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) 
```

### 학습 (다중선형회귀)

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

### 예측값과 실제 값 비교 (테스트 세트)

```python
# 예측 값 
y_pred = reg.predict(X_test)
y_pred
```

```python
# 실제 값
y_test
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델 평가

```python
reg.score(X_train, y_train) # 훈련 세트
```

```python
reg.score(X_test, y_test) # 훈련 세트
```
