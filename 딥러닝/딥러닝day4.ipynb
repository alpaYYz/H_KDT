```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Fullt-connected Network (FC- layer) -> 모두 연결 되어 있기 떄문에 Fullt connected 네트워크이다
class ANN(nn.Module): #class 클래스명(상속받는 class명)
    def __init__(self, input_size, num_classes): #생성자, self는 클래스 안에 있는 함수의 첫번째 입력값
      super(ANN, self).__init__() #nn.Module에서 상속받은 ANN
      self.fc1 = nn.Linear(input_size, 50) # (input = 입력층, 은닉층)
      self.fc2 = nn.Linear(50, num_classes) # (은닉층, 출력층)

    def forward(self, x): # 모델의 활용되는 함수
      x = F.relu(self.fc1(x)) # 입력받은 x의 레이어가 fc1에 들어간다 -> 활성화함수에 적용
      x = self.fc2(x) # 여기는 출력층이라, 활성화 함수 안 들어감
      return x
```

```python
model = ANN(100, 2) # ANN(입력데이터의 컬럼수, 출력층)
```

```python
x = torch.randn(64,100)
print(x.shape)
```

```python
pred = model(x)
pred.shape
```
