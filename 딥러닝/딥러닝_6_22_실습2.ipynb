```python
import torch
import torchvision
import torchvision.transforms as transforms
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_dataset_path = '/content/drive/MyDrive/KDT /3_5_custom_data/training'
test_dataset_path = '/content/drive/MyDrive/KDT /3_5_custom_data/test'
```

```python
train_dataset = torchvision.datasets.ImageFolder(root = train_dataset_path, transform = None)
test_dataset = torchvision.datasets.ImageFolder(root = test_dataset_path, transform = None)
```

```python
for num, value in enumerate(train_dataset):
  data, label = value
  print(num, data, label)
```

```python
train_transforms = transforms.Compose([
    transforms.Resize((100,100)),
    transforms.RandomHorizontalFlip(30),
    transforms.RandomHorizontalFlip(10),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])
# 트레이닝데이터는 학습시키기위해 RandomHorizontalFlip와 RandomHorizontalFlip가 추가된다
```

```python
test_transforms = transforms.Compose([
    transforms.Resize((100,100)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])
```

```python
torchvision.datasets.ImageFolder(root = train_dataset_path, transform = train_transforms)
torchvision.datasets.ImageFolder(root = test_dataset_path, transform = test_transforms)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size = 8, shuffle = True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size = 8, shuffle = False)
```
